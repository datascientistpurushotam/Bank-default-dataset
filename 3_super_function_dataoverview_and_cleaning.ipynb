# Making a Function to give overview of the dataframe

```python
def smart_viewer(dataset):
  #!pip install simple_colors
  #import simple_colors


# This block will show top 10 rows
  print(simple_colors.magenta("\t\t dataset.head(10) Top 10 rows of Dataset "))
  display(dataset.head(10))

# This block will show bottom 10 rows
  print(simple_colors.magenta("\n\t\t dataset.tail(10) Bottom 10 rows of Dataset "))
  display(dataset.tail(10))

# This block is responsible for the info of the dataset
  print(simple_colors.magenta("\n\t\t dataset.info() Compltete info of the Dataset columns "))
  print(dataset.info())

# This block is responsible to see numerical description of dataset columns
  print(simple_colors.magenta("\n\t\t dataset.describe() Numerical summary of Dataset of numerical columns "))
  display(dataset.describe())

# This block is responsible to see number of null values as per column
  print(simple_colors.magenta("\n\t\t dataset.isnull().sum() Number of Null values in each column of dataset "))
  print(simple_colors.blue(dataset.isnull().sum()),'\n\n')

# Plot bar chart of null values
  print(simple_colors.magenta(" \t\t (dataset.isnull().sum()).plot(kind = 'bar') Bar graph plot \n"))
  plt.figure(figsize=(15,7))
  (dataset.isnull().sum()).plot(kind = 'bar')
  plt.show()

# this will give number of unique values
  print('\n',simple_colors.magenta("\n\t\t Number of unique values of Dataset columns "))
  print(simple_colors.blue(dataset.nunique()),'\n') # will give number of unique values

# This block will barplot the number of unique value with column name
  print(simple_colors.magenta(" \t\t (dataset.nunique()).plot(kind = 'bar') Bar graph plot \n"))
  plt.figure(figsize=(15,7))
  (dataset.nunique()).plot(kind='bar')
  plt.show()

# This block will give unique value of each column in List form
  for columnnumber in range(0,len(dataset.columns)):
      print('\n',simple_colors.magenta("\n\t\t Showing unique values in List form of column with details "),'\n',(dataset.iloc[:0,columnnumber]))
      print('\n',dataset.iloc[:,columnnumber].unique(),'\n')

# This block will give unique value of each column in DataFrame form
  for columnnumber in range(0,len(dataset.columns)):
      print('\n',simple_colors.magenta("\n\t\t Showing unique values in Dataframe form of column with details "),'\n',(dataset.iloc[:0,columnnumber]))
      print('\n',dataset.iloc[:,columnnumber].unique,'\n')
```

# Make columns in desired form

## Extracting desired value from term column

```python
datafile['term'] = datafile['term'].str.strip("months")
print(datafile['term'])
```

```python
datafile['emp_length'] = datafile['emp_length'].str.strip('<+ year years')
print(datafile['emp_length'])
```

datafile['emp_length']

datafile['term']

```python
datafile['emp_length'] = datafile['emp_length'].fillna(0)
```

# Drop all null values of the dataset

```python
datafile = datafile.dropna()
```

# Splitting the column To seperate month and year

```python
dataset['issue_d']
```

```python
datafile[['issue_d_month','issue_d_year']] = datafile['issue_d'].str.split("-",expand = True)
```

# Change data type of extracted year to int

```python
datafile['issue_d_year'] = datafile['issue_d_year'].astype('int')
datafile['issue_d_year']
```

```python
datafile['last_pymnt_d_year'] = datafile['last_pymnt_d_year'].astype('int')
```

```python
datafile['last_credit_pull_d_year'] = datafile['last_credit_pull_d_year'].astype('int')
```

```python
datafile['earliest_cr_line_year'] = datafile['earliest_cr_line_year'].astype('int')
```

# Make lambda function to make extracted year ** to ****

```python
datafile['issue_d_year'] = datafile['issue_d_year'].apply( lambda x: x + 2000 if x < 25 else x + 1900)
datafile['issue_d_year']
```

```python
datafile['last_pymnt_d_year'] = datafile['last_pymnt_d_year'].apply( lambda x: x + 2000 if x < 25 else x + 1900)
```

```python
datafile['last_credit_pull_d_year'] = datafile['last_credit_pull_d_year'].apply( lambda x: x + 2000 if x < 25 else x + 1900)
```

```python
datafile['earliest_cr_line_year'] = datafile['earliest_cr_line_year'].apply( lambda x: x + 2000 if x < 25 else x + 1900)
```

# Convert month name to month number

```python
datafile['issue_d_month'] = datafile['issue_d_month'].apply(lambda x : datetime.strptime(x, '%b').month)
datafile['issue_d_month']
```

```python
datafile["last_pymnt_d_month"] = datafile["last_pymnt_d_month"].apply(lambda x : datetime.strptime(x, '%b').month)
```

```python
datafile["last_credit_pull_d_month"] = datafile["last_credit_pull_d_month"].apply(lambda x : datetime.strptime(x, '%b').month)
```

```python
datafile['earliest_cr_line_month'] = datafile['earliest_cr_line_month'].apply(lambda x : datetime.strptime(x, '%b').month)
```

```python
datafile['day'] = 1 # A new dataframe column
```

# concate to make in date format and convert it into datetime

```python
datafile['issue_date'] = pd.to_datetime(datafile["issue_d_year"].astype(str) + '/' + (datafile["issue_d_month"]).astype(str) + '/' + datafile['day'].astype(str))
datafile['issue_date']
```

```python
datafile['last_pymnt_d_date'] = pd.to_datetime(datafile["last_pymnt_d_year"].astype(str) + '/' + (datafile["last_pymnt_d_month"]).astype(str) + '/' + datafile['day'].astype(str))
```

```python
datafile['last_credit_pull_d_date'] = pd.to_datetime(datafile["last_credit_pull_d_year"].astype(str) + '/' + (datafile["last_credit_pull_d_month"]).astype(str) + '/' + datafile['day'].astype(str))
```

```python
datafile['earliest_cr_line_date'] = pd.to_datetime(datafile["earliest_cr_line_year"].astype(str) + '/' + (datafile["earliest_cr_line_month"]).astype(str) + '/' + datafile['day'].astype(str))
```
