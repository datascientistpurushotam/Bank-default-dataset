# Feature Engineering

# VIF Score for Colinear data

but we cannot be implemented it here

for that we have to convert all values to numerical type or we have to do feature engineering

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_score = pd.DataFrame() # creating vif dataframe
vif_score['features'] = dataset.columns #create a column with column names of dataset

# calculating vif score of columns

for col in range(len(dataset.columns)):
  vif_score['score'] = variance_inflation_factor(dataset.values.astype(np.float64), col)


# vif_score['score'] = [variance_inflation_factor (dataset.values.astype(np.float64), i) for i in range(len(dataset.columns)) if np.issubdtype(dataset.iloc[:,col].dtype, np.number) == True]

print(vif_score)
```
